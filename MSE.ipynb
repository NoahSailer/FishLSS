## Setting up the cosmology

```python
# Using halofit to estimate the nonlinear matter power spectrum
mseParams = {
          'output': 'mPk',
          'P_k_max_h/Mpc': 40.,
          'non linear':'halofit', 
          'z_pk': '0.0,10',
          'A_s': 2.10732e-9,
          'n_s': 0.96824,
          'alpha_s': 0.,
          'h': 0.6770,
          'N_ur': 1.0196,
          'N_ncdm': 2,
          'm_ncdm': '0.01,0.05',
          'tau_reio': 0.0568,
          'omega_b': 0.02247,
          'omega_cdm': 0.11923,
          'Omega_k': 0.}

mseCosmo = Class()
mseCosmo.set(mseParams)
mseCosmo.compute()
```

## Creating the experiment
PUMA will observe HI, so I set HI=True. This tells fishlss to use the HI bias and noise models. For now I'm ignoring photo-z errors, so I set $\sigma_z=0$.

```python
mse = experiment(zmin=1.6, zmax=4., nbins=3, fsky=0.24, sigma_z=0.0, MSE=True) #6
```

## Defining the forecast
We will use velocileptors to calculate the nonlinear power spectrum and to model RSD. We choose $k_\text{min}=5\times 10^{-4} {\rm h Mpc^{-1}}$ to be small enough to avoid edge effects in our derivates.

```python
mseCast = fisherForecast(experiment=mse,cosmo=mseCosmo,params=mseParams,khmin=5.e-4,\
                          khmax=1.,Nk=1000,Nmu=200,velocileptors=True,name='MSE')
```

Here we choose the parameters to calculate the derivatives for.

```python
mseCast.marg_params = np.array(['h','log(A_s)','omega_cdm','n_s','omega_b','tau_reio','fEDE','b','N',\
                                 'm_ncdm','N_ur','A_lin','alpha_s','b2',\
                                 'bs','alpha0','alpha2','alpha4','sn2','Omega_k'])
```

## Calculate the derivatives

```python
mseCast.compute_derivatives()
```

```python
mseCast.marg_params = np.array(['A_lin'])
omega_lins = np.logspace(np.log10(10),np.log10(250.),20)
for omega_lin in omega_lins:
   mseCast.omega_lin = omega_lin
   mseCast.compute_derivatives()
```

```python
mseCast.marg_params = np.array(['fEDE'])
log10z_cs = np.linspace(1.5,6.5,20)
for log10z_c in log10z_cs:
   mseCast.log10z_c = log10z_c
   mseCast.compute_derivatives()
```

## Check that the derivatives are reasonable

```python
mseCast.check_derivatives()
```
